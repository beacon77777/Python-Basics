5. What is mutability in Python?
   - Mutability refers to whetherr the contents of an object can be changed after creation.

      - Mutable objects -> can be modified
      - Immutable objects -> can not be changed once created.


```python
# mutable objects (list)
marks = [10,20,15,45,12]
marks[0]=50
print(marks)
```

```python
# Immutable objects (tuple)
marks = (10,20,15,45,12)
marks[2]=50
print(marks)
```

6. Why are lists mutable,but tuples are immutable?
   - list -> Python stores lists as dynamic arrays. This means elements can be added, removed, changed without creating a new object.
   
   - Tuples -> Once created, the memory layout of a tuple is fixed. Python does not allow modification of its elements.
   

```python
## list example
fruits = [ 'Apple','Banana','Dragon Fruits','Orange']
fruits[1] = 'kiwi'
print(fruits)
```

```python
## Tuples example
fruits = ( 'Apple','Banana','Dragon Fruits','Orange')
fruits[1] = 'kiwi'
print(fruits)
```

13. What is the differencce between for and while loops?
    - for loop -> Used when you know the number of iterations in advance.

    - While loop -> Used when you don't know the number of iterations and want to repeat until a condition becomes False.

```python
## for loop example

for i in range(8):
  print(i)
```

```python
## while loop example
num = 0
while num < 5:
  print(num)
  num += 1
```
